# Task 1: Load and Inspect the Data

```python


print('\nTask 1: Load and Inspect the Data')
print('\n')

import pandas as pd
import numpy as np

data = {
    'transaction_id': range(1, 21),
    'date': pd.date_range('2024-10-01', periods=20, freq='D'),
    'region': ['North', 'South', 'East', 'West', None, 'North', 'South', None, 'East', 'West',
               'North', 'South', 'East', 'West', 'North', None, 'East', 'West', 'North', 'South'],
    'product_category': ['Electronics', 'Clothing', None, 'Books', 'Electronics', 'Home',
                         'Clothing', 'Books', 'Electronics', None, 'Home', 'Clothing',
                         'Books', 'Electronics', 'Home', 'Clothing', 'Books', 'Electronics',
                         'Home', 'Clothing'],
    'sales_amount': [1200, 450, 890, None, 1500, 670, None, 340, 2100, 780,
                     560, None, 420, 1800, 920, 510, 380, None, 1100, 640],
    'quantity': [2, 5, 3, 1, None, 4, 2, 3, 1, 5,
                 3, 2, None, 1, 4, 3, 2, 1, None, 4],
    'customer_age': [25, 34, None, 45, 29, None, 38, 52, 27, 41,
                     33, None, 48, 26, 35, 42, None, 31, 39, 44],
    'payment_method': ['Credit Card', 'UPI', 'Cash', 'Debit Card', 'Credit Card',
                       'UPI', 'Cash', None, 'Credit Card', 'UPI', 'Debit Card',
                       'Cash', 'Credit Card', None, 'UPI', 'Cash', 'Debit Card',
                       'Credit Card', 'UPI', None]
}

df = pd.DataFrame(data)

print(df)
print('\n')
```

# Task 1: A) Display basic information using df.info()

```python
# Display basic information using df.info()
print("\nA: Display basic information using df.info()")
print('\n')
print(df.info())
```

#Task 1: B) Count missing values in each column using df.isna().sum()

```python
# Count missing values in each column using df.isna().sum()
print("\nB: Missing Values Count")
print('\n')
df.isnull().sum()
```

# Task 2: Handle Missing Values

```python
# Task 2: 1: Region & Product_category: Fill with mode (most frequent value)
print("\nRegion & Product_category: Filled with mode (most frequent value)")
print('\n')
df['region'].fillna(df['region'].mode()[0])
df['product_category'].fillna(df['product_category'].mode()[0])

# Task 2: 2: Sales_amount: Fill with median
print("\nSales_amount: Filled with median")
print('\n')
df['sales_amount'].fillna(df['sales_amount'].median())

# Task 2: 3: Quantity: Fill using forward fill (ffill)
print("\nQuantity: Filled using forward fill (ffill)")
print('\n')
df['quantity'].fillna(method='ffill')

# Task 2: 4: Customer_age: Fill with mean (rounded to integer)
print("\nCustomer_age: Filled with mean (rounded to integer)")
print('\n')
df['customer_age'].fillna(round(df['customer_age'].mean()))

# Task 2: 5: Payment_method: Drop rows where payment_method is missing
print("\nPayment_method: Dropped rows where payment_method is missing")
print('\n')
df.dropna(subset=['payment_method'])

# Verify no missing values remain using df.isna().sum()
print("\nVerify no missing values remain using df.isna().sum()")
print('\n')
print(df.isnull().sum())
```

# Task 3: Group By Analysis

```python
# Task 3: 1: Calculate total sales by region
print("\nTask 3: 1: Calculate total sales by region")
print('\n')
total_sales_by_region = df.groupby('region')['sales_amount'].sum()
print(total_sales_by_region)
print('\n')

# Task 3: 2: Calculate average sales by product_category
print("\nTask 3: 2: Calculate average sales by product_category")
print('\n')
average_sales_by_product_category = df.groupby('product_category')['sales_amount'].mean()
print(average_sales_by_product_category)
print('\n')

# Task 3: 3: Group by both region and product_category, calculate total sales and quantity, then use reset_index() to flatten
print("\nTask 3: 3: Group by both region and product_category, calculate total sales and quantity, then use reset_index() to flatten")
print('\n')

grouped_df = df.groupby(['region', 'product_category']).agg({
    'sales_amount': 'sum',
    'quantity': 'sum'
}).reset_index()

print(grouped_df)
print('\n')

# Task 3: 4: Display top 3 region-product combinations by sales
print("\nTask 3: 4: Display top 3 region-product combinations by sales")
print('\n')
top_3_regions_products = grouped_df.nlargest(3, 'sales_amount')
print(top_3_regions_products)
print('\n')
```

# Task 4: Custom Aggregation

```python
# Task 4: 1: Create a function sales_range() that returns max - min of sales
print("\nTask 4: 1: function sales_range() is created that returns max - min of sales")
print('\n')
def sales_range(x):
    return x.max() - x.min()

# Task 4: 2: Apply it to find sales range for each region
print("\nTask 4: 2: Apply it to find sales range for each region")
print('\n')
sales_range_by_region = df.groupby('region')['sales_amount'].agg(sales_range).reset_index()
print(sales_range_by_region)
print('\n')

# Task 4: 3: Use .agg() to calculate multiple metrics by region:
#sales_amount: sum, mean, max
#quantity: sum, min

print("\nTask 4: 3: Use .agg() to calculate multiple metrics by region")
print('\n')
custom_agg_by_region = df.groupby('region').agg({
    'sales_amount': ['sum', 'mean','max'],
    'quantity': ['sum', 'min']
}).reset_index()
print(custom_agg_by_region)
print
```
